# Data Preprocessing
In the data preprocessing step, we clarify the data types of the columns we need and take care that the data is very clean and not missing values.

## Data Manipulation

```python
data_frame.isnull().sum() # Is there any null value?
```

Item_Fabric_Amount

```python
data_frame['Item_Fabric_Amount'].fillna(data_frame['Item_Fabric_Amount'].mean(), inplace=True)
# we filled the null variables with the mean so we don't affect the general mean
```

Store_Size

```python
# Let's match the most common value in the "Store_Size" column for each unique "Store_Type" value
mode_of_store_size = data_frame.pivot_table(values='Store_Size', columns='Store_Type', aggfunc=lambda x: x.mode()[0])
mode_of_store_size
```

```python
miss_values = data_frame['Store_Size'].isnull()
miss_values.value_counts()
```

```python
data_frame.loc[miss_values, 'Store_Size'] = data_frame.loc[miss_values, 'Store_Type'].apply(lambda x: mode_of_store_size[x])
# Null values are replaced with the corresponding mode value for each "Store_Type" value. This ensures that missing data is properly completed.
```

## Data Cleaning

## Data Conversion
Machine learning algorithms generally work with numerical data. Numerical data generally requires less memory space and is more computationally efficient than categorical data. The best way to do this is to use LabelEncoder.

```python
# required libraries
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
data_frame['Store_Id'] = encoder.fit_transform(data_frame['Store_Id'])
data_frame['Store_Size'] = encoder.fit_transform(data_frame['Store_Size'])
data_frame['Store_Location_Type'] = encoder.fit_transform(data_frame['Store_Location_Type'])
data_frame['Store_Type'] = encoder.fit_transform(data_frame['Store_Type'])

data_frame['Item_Id'] = encoder.fit_transform(data_frame['Item_Id'])
data_frame['Item_Fit_Type'] = encoder.fit_transform(data_frame['Item_Fit_Type'])
data_frame['Item_Fabric'] = encoder.fit_transform(data_frame['Item_Fabric'])
```

```python
# displaying label encoded dataset
data_frame
```

## Data Splitting (Test and Train)
We have to separate the data into features and target variable. We will then split the Data into training and test sets.

```python
# required libraries
from sklearn.model_selection import train_test_split
```

```python
X = data_frame.drop(columns='Item_Store_Sales', axis=1)
X.shape
```

```python
Y = data_frame['Item_Store_Sales']
Y.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

# Machine Learning Model Training

```python
# common libraries
from sklearn import metrics # for modelling evaluation
import pickle # for saving and loading models
```

### randomForest Regressor

```python
# required libraries
from sklearn.ensemble import RandomForestRegressor
regressor_randomForest = RandomForestRegressor()
```

```python
# Machine Learning Model Training
regressor_randomForest.fit(X_train, Y_train)
y_pred_randomForest = regressor_randomForest.predict(X_test)
```

```python
# Model Evaluation
print("R-squared Score:", metrics.r2_score(Y_test, y_pred_randomForest))
```

```python
# Saving trained model
pickle.dump(regressor_randomForest, open('models/model_randomForest.pkl','wb'))
print("Saved model")
```

### XGBoost Regressor

```python
# required libraries
from xgboost import XGBRegressor
regressor_xgboost = XGBRegressor()
```

```python
# Machine Learning Model Training
regressor_xgboost.fit(X_train, Y_train)
y_pred_xgb = regressor_xgboost.predict(X_test)
```

```python
# Model Evaluation : Metrics for Continuous Target Regression
print("R-squared Score:", metrics.r2_score(Y_test, y_pred_xgb))
```

```python
# Saving trained model
pickle.dump(regressor_xgboost, open('models/model_xgboost.pkl','wb'))
print("Saved model")
```

### DecisionTree Regressor

```python
# required libraries
from sklearn.tree import DecisionTreeRegressor
regressor_decisionTree = DecisionTreeRegressor()
```

```python
# Machine Learning Model Training
regressor_decisionTree.fit(X_train, Y_train)
y_pred_decisionTree = regressor_decisionTree.predict(X_test)
```

```python
# Model Evaluation
print("R-squared Score:", metrics.r2_score(Y_test, y_pred_decisionTree))
# negative value is the result of an unsuccessful modeling
```

```python
# Saving trained model
pickle.dump(regressor_decisionTree, open('models/model_decisionTree.pkl','wb'))
print("Saved model")
```

## Model Comparison

```python
# Model Comparison with Heatmap
models = ['RandomForestRegressor', 'XGBRegressor', 'DecisionTreeRegressor']
r2_scores = [metrics.r2_score(Y_test, y_pred_xgb), metrics.r2_score(Y_test, y_pred_randomForest), metrics.r2_score(Y_test, y_pred_decisionTree)]

plt.figure(figsize=(9,6)) # for monitoring 3 parameters
sns.heatmap(np.array(r2_scores).reshape(1, -1), annot=True, cmap='Blues', yticklabels=True, cbar=True)
plt.xticks(ticks=np.arange(len(models))+0.5, labels=models)
plt.yticks([])
plt.title('Model Comparison - R-squared Score')
```
